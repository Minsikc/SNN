```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#  
n_out = [2, 10, 30, 50, 100, 200]
bptt_values = [0, 0.04, 0.072, 0.10, 0.123, 0.156]
et_values = [0, 0.148, 0.234, 0.313, 0.335, 0.35] #     
user_et_values = [0, 0.08, 0.145, 0.232, 0.224, 0.256] #    

#  
data = {
    'n_out': n_out * 2,
    'normalized distance': bptt_values + user_et_values,
    'Algorithm': ['BPTT'] * len(n_out) + ['ET Algorithm'] * len(n_out)
}
df = pd.DataFrame(data)

#   
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Seaborn   
ax = sns.lineplot(
    data=df,
    x='n_out',
    y='normalized distance',
    hue='Algorithm',
    style='Algorithm',
    markers=['o', 's'],
    markersize=8,
    palette=['blue', 'red'],
    dashes=[(1, 0), (4, 2)] # BPTT , ET 
)

#    
plt.title('BPTT vs. ET Algorithm Performance', fontsize=16)
plt.xlabel('n_out', fontsize=15)
plt.ylabel('normalized distance', fontsize=15) # y  

#     
plt.legend(title='Algorithm', loc='lower right')

#  
plt.show()
```
